```python
import numpy as np
import pandas as pd

tennis = pd.read_csv('PlayTennis.csv')
tennis
```

```python
ct_outlook=pd.crosstab(tennis['Outlook'], tennis['Play Tennis'], margins = True)
ct_outlook
```

```python
ct_temp=pd.crosstab(tennis['Temperature'], tennis['Play Tennis'], margins = True)
ct_temp
```

```python
ct_humidity=pd.crosstab(tennis['Humidity'], tennis['Play Tennis'], margins = True)
ct_humidity
```

```python
ct_windy=pd.crosstab(tennis['Wind'], tennis['Play Tennis'], margins =True)
ct_windy
```

```python
ct_play=pd.crosstab(index=tennis['Play Tennis'],columns="count", margins=True)
ct_play
```

```python
p_yes = ct_play.iloc[1,0] /ct_play.iloc[2,0]* ct_outlook.iloc[2,1]/ct_outlook.iloc[3,1]* ct_temp.iloc[0,1]/ct_temp.iloc[3,1]*ct_humidity.iloc[0,1]/ct_humidity.iloc[2,1]* ct_windy.iloc[0,1]/ct_windy.iloc[2,1]
print('The probability of tennis played with given conditions is', p_yes)

p_no = ct_play.iloc[0,0] /ct_play.iloc[2,0]* ct_outlook.iloc[2,0]/ct_outlook.iloc[3,0]* ct_temp.iloc[0,0]/ct_temp.iloc[3,0]*ct_humidity.iloc[0,0]/ct_humidity.iloc[2,0]*ct_windy.iloc[0,0]/ct_windy.iloc[2,0]
print('The probability of tennis not played with given conditions is','%.3f'%p_no)
```

```python
#MAP
if p_yes > p_no:
  print('Tennis match will be conducted when the outlook is sunny,the temperature is hot, there is high humidity and windy is higher')
else:
  print('Tennis match will not be conducted when the outlook is sunny,the temperature is hot, there is high humidity and windy is higher')
```
